```python
#clean up data replace month number with associated month name
weather_df['Month'] = weather_df['Month'].replace(
    {1:'January',2:'February',3:'March',4:'April',5:'May',6:'June',7:'July',
     8:'August',9:'September',10:'October',11:'November',12:'December'})

#remove unwanted columns
weather = weather_df[['Year','Month', 'Condition','Day']]

#Rename Columns
WT = weather.rename(columns={"Year":"Year", "Month":"Month","Condition":"Condition","Day":"Condition_Count"})

#Group weather dataframe and calcuate weather conditions per month per year
WT2 = WT.groupby(['Year','Month','Condition']).count()

#Reset Index
weather_table = WT2.reset_index()
weather_table.head()
```

```python
#covert dataframe to CSV file
weather_table_df = weather_table.to_csv (r'C:\Users\fenel\Turkish_Airlines_AF_BW\Project\weather_table.csv'
               , index = None, header=True)
```

```python
#clean up data replace month number with associated month name
stock_df['Month'] = stock_df['Month'].replace(
    {'Ocak':'January','Şubat':'Febeyary','Mart':'March','Nisan':'April','Mayıs':'May','Haziran':'June','Temmuz':'July',
     'Ağustos':'August','Eylül':'September','Ekim':'October','Kasım':'November','Aralık':'December'})

#remove unwanted columns and readjust columns
stock = stock_df[['Year','Month', ' Last Price', ' Lowest Price',
       ' Highest Price', ' Volume']]

#Rename Columns
rename_stock = stock.rename(columns={"Year":"Year", "Month":"Month"," Last Price":"Avg. Close",
                                           " Lowest Price":"Avg Low"," Highest Price":"Avg High",
                                           " Volume":"Avg Volume"})

#Group stock dataframe and calcuate average stock per month per year
stock_df_group = rename_stock.groupby(['Year','Month']).mean()

#Reset Index
stock_table = stock_df_group.reset_index()

#format columns
stock_table['Avg. Close'] = stock_table['Avg. Close'].map("{:.2f}".format)
stock_table['Avg Low'] = stock_table['Avg Low'].map("{:.2f}".format)
stock_table['Avg High'] = stock_table['Avg High'].map("{:.2f}".format)
stock_table['Avg Volume'] = stock_table['Avg Volume'].map("{:.2f}".format)
stock_table.head()
```

```python
#covert dataframe to CSV file
stock_table_df = stock_table.to_csv (r'C:\Users\fenel\Turkish_Airlines_AF_BW\Project\stock_table.csv'
               , index = None, header=True)
```
